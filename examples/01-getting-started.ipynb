# Getting Started with Sondio

Install the SDK, plug in your API key, and run your first query — all in under a minute.

## 1. Install

```bash
pip install sondio           # core (DataFrames only)
pip install 'sondio[geo]'    # adds GeoDataFrame support
```

## 2. Set your API key

Grab one at [sondio.io/developers](https://sondio.io/developers). Then set it however you like — in-notebook, via environment variable, or in `~/.sondio/config`.

````python
import os
import sondio

sondio.api_key = os.environ.get("SONDIO_API_KEY", "sk_sondio_your_key_here")
print(f"sondio {sondio.__version__}")
````

## 3. Your first query

Every call returns a pandas DataFrame. Here are recent magnitude-3+ earthquakes in Texas:

````python
quakes = sondio.earthquakes(state="TX", min_mag=3.0, days=90, limit=20, page=1)
quakes.head()
````

Pagination is opt-in. The line above fetched page 1 only — pass `all_pages=True` to iterate or `page=N` for a specific page.

## 4. Other namespaces

The SDK mirrors the Sondio [naming convention](https://github.com/sondio-io/sondio/blob/main/docs/reference/dataset-naming.md):

| Call | Dataset |
|------|---------|
| `sondio.oilgas.wells(country="US", state="TX")` | Oil & gas wells |
| `sondio.oilgas.permits(state="TX")` | Drilling permits |
| `sondio.us.ghg.facilities(state="TX")` | EPA GHGRP facilities |
| `sondio.us.npdes.permits(state="TX")` | Clean Water Act permits |
| `sondio.us.phmsa.pipeline_incidents(state="TX")` | Pipeline incidents |
| `sondio.us.epa.aquifer_exemptions(state="TX")` | UIC aquifer exemptions |
| `sondio.earthquakes(state="TX")` | USGS earthquakes |
| `sondio.geo.subdivisions(country="US")` | States / provinces |

````python
facs = sondio.us.ghg.facilities(state="TX", limit=10, page=1)
facs[["facility_name", "city", "industry_type", "total_co2e"]].sort_values("total_co2e", ascending=False)
````
